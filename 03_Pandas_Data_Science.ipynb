<h2 style="color:#c4960c">Arrests for Marijuana Possession</h2>
<i><p>Below is how to import .csv files into a panda DataFrame</p></i>

```python
df = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/carData/Arrests.csv', header=0)
df.head()
```

<br><br>
<hr>
<h3 style="color:#c4960c"><i>Checks by Age and Sex</i></h3>
<p>Below are some examples of creating Pivot Tables and Charts</p>

<h4>Pivot Table</h4>

```python
pivot = df.pivot_table(values=["checks"], aggfunc=np.sum, index=["age"], columns=["sex"])
#pivot
```

```python
pivot.to_csv('Arrests for Marijuana Possession.csv')
```

<h4>Chart</h4>
<i><p>This is a histagram style bar char (not a true histagram)</p><i>

```python
pt = pivot.plot.bar(figsize=(20,10))
pt.grid()
plt.savefig('Arrests for Marijuana Possession.png')
```

<br><br>

<h3 style="color:#c4960c"><i>Checks by Age, Colour and Sex</i></h3>

<h4>Pivot Table</h4>

```python
pivot = df.pivot_table(values=["checks"], aggfunc=np.sum, index=["year"], columns=["sex", "colour"])
pivot
```

<h3 style="color:#c4960c"><i>Checks by Age, Colour and Year</i></h3>

<h4>Pivot Table</h4>
<i><p>Below are the main elements of a pivot table:</p>
<li>values=["column_name", "column_name"]</li>
<li>index=["column_name", "column_name"]</li>
<li>columns=["column_name", "column_name"]</li>
<li>aggfunc=np.sum or: 
    <ul>np.max</ul>
    <ul>np.min</ul>
    <ul>np.mean</ul>
    <ul>....</ul>
</li></i>

```python
pivot = df.pivot_table(values=["checks"], aggfunc=np.sum, index=["year"], columns=["sex", "colour"])
pivot
```

<h4>Chart</h4>
<i><p>This is a line/plot chart</p></i>

```python
pt = pivot.plot(figsize=(20,10),title="Arrests for Marijuana Possession")
pt.grid()
```

<i><p>This is a density chart</p></i>

```python
pt = pivot.plot.density(figsize=(20,10),title="Arrests for Marijuana Possession")
pt.grid()
```

 <i><p>This is a box chart</p></i>

```python
pt = pivot.plot.box(figsize=(20,10))
pt.grid()
```

```python
pivot = df.pivot_table(values=["checks"], aggfunc=np.sum, index=["year"])
Pivot = pd.DataFrame(pivot.to_records())
Pivot
```

```python
y = Pivot['checks']
x = Pivot['year']
```

<i><p>Below is a list, map and lambda function converting each item in the 'x'(year) list from a number to a string</p></i>

```python
toStr = lambda x:  str(x)
x = list(map(toStr, x))
x #Years as aa list of trings
```
